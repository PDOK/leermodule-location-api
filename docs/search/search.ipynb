# PDOK Search Notebook
Interactieve voorbeelden voor het `/search` endpoint.

```python
import requests, json, html

def fetch(url):
    url = html.unescape(url)
    r = requests.get(url)
    try: return r.json()
    except: return r.text
```

# Zoeken met het `/search` endpoint

Het `/search` endpoint ondersteunt tekstgebaseerd zoeken over meerdere collecties.

## Minimale query

Een eenvoudige zoekopdracht vereist alleen een zoekterm (`q`-parameter), maar geeft een opmerkelijk resultaat:



```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=am')
print(json.dumps(res, indent=2, ensure_ascii=False))
```



Wat betekend de response en wat moet je voor "foo" invullen?

De API dwingt je af om bewust te kiezen welke collecties je wilt doorzoeken. Zonder duidelijke keuze retourneert het endpoint geen resultaten. Dit beschermt tegen ongewilde zoekopdrachten over alle gegevens. Bovendien kan PDOK makkelijk nieuwe collecties toevoegen aan de API zonder dat jouw bestaande zoekopdrachten veranderen.

## Collecties activeren

Standaard zoekt het endpoint niet in alle beschikbare collecties. Een collectie wordt pas actief gebruikt in zoekopdrachten wanneer je een collection inclusief `version` toevoegt.

Voorbeeld: alleen adressen activeren:



```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=Dam&adres[version]=1')
print(json.dumps(res, indent=2, ensure_ascii=False))
```



Geeft ook een opmerkelijk resultaat. Gebruik `--globoff` zodat Curl de `[` en `]` goed kan interpreteren:



```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=Dam&adres[version]=1')
print(json.dumps(res, indent=2, ensure_ascii=False))
```



Dit zoekt alleen in de adrescollectie naar "Dam".

Meerdere collecties tegelijk:



```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=Rot&adres[version]=1&woonplaats[version]=1')
print(json.dumps(res, indent=2, ensure_ascii=False))
```



Nu worden zowel adressen als woonplaatsen doorzocht.

Zoek uit welke collecties je kan gebruiken in de zoek opdracht



```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/collections')
print(json.dumps(res, indent=2, ensure_ascii=False))
```


Welke collecties worden het meest gebruikt en waarom?

## Ranking met relevance

Je kunt de zoekresultaten beïnvloeden door prioriteiten in te stellen per collectie met de `relevance`-parameter:



```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=Maas&adres[version]=1&adres[relevance]=0.7&woonplaats[version]=1&woonplaats[relevance]=0.3')
print(json.dumps(res, indent=2, ensure_ascii=False))
```



Voorbeeld met verschillende prioriteiten:



```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=utrecht&gemeentegebied[relevance]=0.1&gemeentegebied[version]=1&provinciegebied[relevance]=0.9&provinciegebied[version]=1&woonplaats[relevance]=0.1&woonplaats[version]=1&f=json')
print(json.dumps(res, indent=2, ensure_ascii=False))
```

```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=utrecht&gemeentegebied[relevance]=0.9&gemeentegebied[version]=1&provinciegebied[relevance]=0.1&provinciegebied[version]=1&woonplaats[relevance]=0.1&woonplaats[version]=1&f=json')
print(json.dumps(res, indent=2, ensure_ascii=False))
```

```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=utrecht&gemeentegebied[relevance]=0.1&gemeentegebied[version]=1&provinciegebied[relevance]=0.1&provinciegebied[version]=1&woonplaats[relevance]=0.9&woonplaats[version]=1&f=json')
print(json.dumps(res, indent=2, ensure_ascii=False))
```



## Bbox-filtering



```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=Arn&bbox=5.85,51.95,5.90,52.00&bbox-crs=http://www.opengis.net/def/crs/OGC/1.3/CRS84&woonplaats[version]=1')
print(json.dumps(res, indent=2, ensure_ascii=False))
```



## Andere CRS terugvragen



```python
res = fetch('https://api.pdok.nl/kadaster/location-api/v1-demo/search?q=Onze+Lieve+Vrouwetoren&crs=http://www.opengis.net/def/crs/EPSG/0/28992&gebouw[version]=1')
print(json.dumps(res, indent=2, ensure_ascii=False))
```



## Complete uitgebreide query
