# Oefening: Basiszoekopdrachten in de PDOK Location API
Gebruik requests om een eenvoudige zoekopdracht uit te voeren en bekijk de response.

```python
import requests, json
url='https://api.pdok.nl/kadaster/location-api/v1-demo/search'
params={'q':'Dam','adres[version]':1}
r=requests.get(url, params=params)
r.raise_for_status()
print('URL:', r.url)
print('Status:', r.status_code)
print(json.dumps(r.json(), indent=2)[:2000])
```
