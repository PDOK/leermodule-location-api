# Oefening: BBOX-filtering in de PDOK Location API
Beperk de zoekresultaten met een bounding box.

```python
import requests, json
url='https://api.pdok.nl/kadaster/location-api/v1-demo/search'
params={'q':'Arn','bbox':'5.85,51.95,5.90,52.00','bbox-crs':'http://www.opengis.net/def/crs/OGC/1.3/CRS84','adres[version]':1}
r=requests.get(url, params=params)
r.raise_for_status()
print('URL:', r.url)
print(json.dumps(r.json(), indent=2)[:2000])
```
